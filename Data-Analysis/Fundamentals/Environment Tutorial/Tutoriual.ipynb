# How to set up a new Python Project

If you're about to set up your very first project in python, this tutorial is just for you. We will explain what software you need to download and what steps to take to successfully create a new python project and to correctly set it up on Gitlab so that you can share your code with your colleagues.

First, let's check whether you already have all the necessary installations and requirements. Most importantly, you need to install an editor / IDE to run Code in your programming language. Most of us use `Visual Studio Code` because it is free, user-friendly and runs on all the common operating systems. If you haven't downloaded it yet, you can do so now: https://code.visualstudio.com/Download (choose the respective downloading option for your operating system).

## 2. Clone your project

Now that you have created your project on Gitlab, you  need to `clone` it in VS Code. To do so, go to the source control tab on the menu bar and click "Clone Repository".

A box will open up where you are asked to enter your gitlab key. You will find it on Gitlab when you maneuver to your newly created project and click on the blue button that says "Code". There you will see two options to copy a URL to clone your repository with, with SSH or with HTTPS. Assuming you have already set up and connected your SSH keys, we will choose that one. This will connect your project to Gitlab via git and make it much easier for you to update your process to Gitlab. So we simply click on the button next to the URL 

<img src="Copy.png"  width="300" height="70"/> 

and enter it in the field that opened in VS Code. You will then get asked to provide a parent folder on your hard drive for your project. It is recommended you create a folder on your desktop where all your future Gitlab projects will live. If you choose this folder, a parent folder named after your project will automatically be created inside the the folder you chose. After you click "Ok", you have  connected VS Code to your project on Gitlab. You are going to be asked whether you want to open the cloned repository. Click "Open". You are now ready to prepare the environment for your new project. 

## 3. Add .gitignore
For every new project that you create, there are some files you will always need. 
The first one that we are going to add is `.gitignore`. The easiest way is to create a new file and to name it `.gitignore`. This will automatically create the correct file. The purpose of this file is to tell git which files it should ignore when you push your changes to Gitlab. When dealing with a large amount of data for your analysis, syncing all of it to Gitlab would be too much for the storage space (which is limited to 10GB per repository). This is why it is very important to define the type of files or specific folder names that you want to keep local only. Here you can see an example of how to exclude all meg- (.fif) and mat-files or whole folders from bein pushed to Gitlab:

```
*.fif
*.mat
yourfolder
```


## 4. Set up your environment
The next step before you can actually start writing and using your code is to set up an environment for your project. You may wonder what an environment is - essentially it contains all the necessary packages as well as your python interpreter that allows your code to run. Remember that each project has its own custom environment that you set up depending on the specific stuff you plan to do.

In our lab we use `pixi`to manage the environment for us. It handles all the dependencies and integrates them in a central folder. If you haven't installed pixi yet, it is time to do so: 

Run this line in your terminal for Linux or MacOS systems:

```curl -fsSL https://pixi.sh/install.sh | sh ```


And this line for Windows:

```powershell -ExecutionPolicy ByPass -c "irm -useb https://pixi.sh/install.ps1 | iex"```  or download the installer via the following link: https://pixi.sh/latest/#__tabbed_1_2 

Once you downloaded pixi, you initialize it in your current python project by running `pixi init` in your terminal. This will create a file called `pixi.toml` in your projects repository. If you open it you will see something like this:
```
[project]
authors = ["YourGitlabHandle <max.mustermann@plus.ac.at>"]
channels = ["conda-forge"]
name = "python-project"
platforms = ["osx-arm64"]
version = "0.1.0"

[tasks]

[dependencies]
```

Every toolbox you want to use must first be added via pixi. This is done by running the command `pixi add <nameoftoolbox>` in your terminal. This will automatically update your `pixi.toml` file. Go try it out and copy the following command:

`pixi add python`

Your pixi.toml file should now list python under the dependencies. Note that the version number of the dependency is always listed next to the name. Sometimes you may need a specific version because other packages you have installed rely on earlier or later versions. You can simply modify it in the pixi.toml directly or specify the version number when adding the package like so:

`pixi add python = 3.12`

If you manually alter the pixi.toml file, make sure to run `pixi install` after, to install the environment. There are many more commands and specific usage cases for pixi which you can learn more about here: https://pixi.sh/latest/.

Before running python code in your project, make sure the pixi environment is activated by running `pixi shell` in your terminal. You do not need to do this if you're using jupyter, which we will discuss in a minute.

## 5. Run some code

Okay, now let's run some code. Create a new python file (it is automatically recognized as a python file if you give it a name that ends in .py). First, you always need to import the packages that you use in your script, e.g.:

`````python
#%% Import the packages you need
import mne # not needed in this example but most of the time you will need it for your analysis of brain data
import numpy as np
import pandas as pd
`````

Notice how the block starts with `#%%`? Adding that at the beginning of your code block marks sections that can be run separately. You can either run your code via a python terminal or via an interactive window using a software called jupyter. In order to use it, you need to install the Jupyter Extension and add jupyter to your dependencies (pixi add ipykernel). Using interactive windows is recommended because it allows you to plot what you run and to interact with it. So if you click `Run cell` above your code section, an interactive window will open automatically. Now you are ready to play around with some more code:

`````python
#%% Create a dataframe
columns = ['Subject_ID', 'Course_ID', 'Grade']
data = [
    (103, 201, 1),
    (103, 203, 2),
    (103, 204, 3)  
]

df = pd.DataFrame(data, columns=columns)

mean_grade = np.mean(df['Grade'])

# Add a row to display the mean grade
df.loc['Mean'] = ['-', '-', mean_grade]

print(df)
`````

Now that you have written your first code snippet, make sure to save your file. And most importantly, make it a habit to sync your progress regularly to Gitlab to prevent any loss. You can either do this via the Source control button that you have already learned about when cloning the repository or via the terminal, using git commands. The easiest and most straightforward option is to navigate to the Source control menu and to select the files and changes you want to commit. Usually you want to commit all the work you have done so we are going to click the `+` that appears when you hover your mouse over the "Changes" button. But if you only want to commit single files, you can simply navigate to it and click the `+` that appears on said file. Notice how your files are now listed under a button saying "Staged Changes" instead and the list under "Changes" is empty? Now your files/changes are ready to be commited. But first you should enter an informative commit message that will allow you to retrace what you have accomplished that day in case you need to go back to earlier versions of your code. After you have committed your changes, another button is going to appear where you can `Sync` your changes, which is going to push them to Gitlab. You can check this by opening your project on Gitlab and refresh the page. Your changes should now be visible and your commit message should be written next to all the files that you modified the last time you pushed your changes. 

If you are collaborating with a colleague on a project and you both work on the same Gitlab project, you should always make sure to run the following command before you start working on the project:

```
git fetch
git pull
````

By doing this, you make sure that your local repository is up to date with all the changes that your colleague has made. Note, that sooner or later you are probably going to run into a merge conflict - don't panic. If you have regularly synced your changes to Gitlab you cannot loose your progress. So just make sure you understand where the conflict is coming from and to resolve it. You will find many tutorials on how to handle your merge conflict online, e.g.: https://blog.stackademic.com/git-rebase-explained-like-youre-new-to-git-263c19fa86ec, or https://www.geeksforgeeks.org/git/merge-conflicts-and-how-to-handle-them/.
